### Import all packages that is needed

```python
import wfdb
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import os
import shutil
import pandas as pd
import matplotlib.pyplot as plt
```

### Look for wich database you want to use

```python
dbs = wfdb.get_dbs()
display(dbs)
```

### In the next cell we will download recordings from https://physionet.org/physiobank/database/nsrdb/ without storing it localy on the PC.

```python
record = wfdb.rdrecord('16265', pb_dir='nsrdb/')
wfdb.plot_wfdb(record=record, title='Record 16265 from Physionet NSRDB') 
print(record.__dict__)
```

### A whole database can be downloaded with this command:

```python
wfdb.dl_database('afdb', 'data/afdb')
```

### To import data that is easier to work with you can do like this:  (from a local folder)

```python
signals, fields = wfdb.rdsamp('data/nsrdb/16265', sampfrom=800)
```

### This is what you get from "fields" and "signals":

```python
plt.plot(signals)
```

```python
fields
```

### Since "signals" is a numpy array we can transform it to a pandas dataframe and select the signal/column we want to use

```python
type(signals)
```

```python
ecg_df=pd.DataFrame(signals)
```

```python
plt.plot(ecg_df[0])
```

### To be able to see any ECG-patterns we have to select a shorter range of the signals

```python
smal_ecg=ecg_df[0].iloc[0:500]
```

```python
plt.plot(smal_ecg)
```
